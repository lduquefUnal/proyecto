```python
import boto3
s3 = boto3.client('s3')
s3.upload_file(
    Filename="proyecto/modelos/mnist_quantum/model.tar.gz",
    Bucket='modelos-sagemaker',
    Key='modelos/mnist/mnist_quantum/model.tar.gz'
)
```

```python
from sagemaker.pytorch import PyTorchModel
import sagemaker

role = sagemaker.get_execution_role()

model = PyTorchModel(
    model_data='s3://modelos-sagemaker/modelos/mnist/mnist_quantum/model.tar.gz',
    role=role,
    entry_point='inference.py',
    source_dir='code',
    framework_version='2.0',
    py_version='py310'
)
```

```python
predictor = model.deploy(
    endpoint_name='mnist-quantum-endpoint',
    instance_type='ml.m5.large',   # o la instancia que necesites
    initial_instance_count=1
)
```
